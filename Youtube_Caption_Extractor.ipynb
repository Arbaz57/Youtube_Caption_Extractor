```python
from youtube_transcript_api import YouTubeTranscriptApi
def youtube_caption_extractor(vid_id):
  srt = YouTubeTranscriptApi.get_transcript(vid_id)
  text = []
  for i in srt:
    s = i["text"]
    text.append(s)
  captions = ""
  for i in text:
    captions+= i
    captions+= " "
  return captions
```

```python
captions = youtube_caption_extractor("d7sUWwHugg8")
```
